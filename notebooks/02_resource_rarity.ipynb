# Rarity of Resources

```python
import json
import pandas as pd
import seaborn as sns
```

```python
sns.set_theme(style="white", palette="rocket")
```

```python
with open("../data/resources.json") as f:
    resources = json.load(f)
```

```python
resources
```

```python
df = pd.DataFrame(resources)
df
```

```python
# normalize to lowest
iron_rarity = 3.579805
df["rarity"] = df["rarity"].map(lambda r: r / iron_rarity)
df
```

```python
sns.barplot(
    data=df.sort_values("percentage_of_total", ascending=False),
    x="percentage_of_total",
    y="slug",
    orient="y",
)
```

```python
sns.barplot(
    data=df.sort_values("rarity", ascending=False), x="rarity", y="slug", orient="y"
)
```
